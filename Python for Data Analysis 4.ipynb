#### Using ```loc``` and ```iloc``` to select rows and columns in a dataframe

````python
### df.loc[row_label_range/value, column_label_range/value]
### df.iloc[row_index_range/value, column_index_range/value]
````

````python
df_hospital_appts.loc[:,:]
````

````python
df_hospital_appts.loc[2,'ScheduledDay']
````

````python
df_hospital_appts.loc[:100,:'Scholarship']
````
