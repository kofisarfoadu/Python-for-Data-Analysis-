### Defining functions
Functions are reusable code blocks that you can use to perform a single action.

````python
def print_argument(argument):
    """this function prints argument"""  ## this docstrings
    print(argument)
````

````python
print_argument('Great Success')
````

````python
def return_argument(argument):
    return argument
````

````python
return_argument('Great Success')
````

#### Using the ```apply``` function on Dataframes

````python
## Read about the range function here:  https://www.w3schools.com/python/ref_func_range.asp
age_65_to_150 = list(range(65, 151, 1))
````

````python
print(age_65_to_150)
````

````python
def f_age_65_to_150(age):
    if age in age_65_to_150:
        return 1
    else:
        return 0
        
````

````python
f_age_65_to_150(65)
````

````python
f_age_65_to_150(64)
````

````python
df_hospital_appts['Age']
````

````python
df_hospital_appts['Age_65_and_over'] = df_hospital_appts['Age'].apply(f_age_65_to_150)
````

````python
df_hospital_appts.head()
````

````python
df_hospital_appts.tail()
````

````python
df_hospital_appts['Age_65_and_over'].value_counts()
````
