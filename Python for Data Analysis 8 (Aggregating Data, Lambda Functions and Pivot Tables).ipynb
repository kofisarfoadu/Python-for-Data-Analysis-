Aggregating Using ```groupby``` method

````python
import pandas as pd
````

````python
df_hospital_appts = pd.read_csv('noshowappointments-kagglev2-may-2016.csv')
````

````python
df_hospital_appts.info()
````

#### Take a sample of data (maybe 10 samples from the dataset)

````python
df_hospital_appts.sample(n=10)
````

#### Sort values ```sort_values()```
##### Lets sort the records by age

````python
df_hospital_appts.sort_values(by='Age', ascending=False)
````

````python
df_hospital_appts.sort_values(by='ScheduledDay', ascending=False)
````

#### Using inline functions or ```lambda``` functions

````python
def f_age_over_65(age):
    if age > 65:
        return 1
    else:
        return 0
````

````python
df_hospital_appts['Age']
````

````python
df_hospital_appts['Age_Over_65'] = df_hospital_appts['Age'].apply(f_age_over_65)
````

````python
df_hospital_appts.sample(n=20)
````

````python
df_hospital_appts['Over_65_Lambda'] = df_hospital_appts['Age'].apply(lambda x: x > 65)
````

````python
df_hospital_appts.sample(n=12)
````

#### Group data by categorical Variables

````python
group_by_gender = df_hospital_appts.groupby(['Gender'])
````

````python
group_by_neighborhood = df_hospital_appts.groupby(['Neighbourhood'])
````

````python
group_by_gender_neighborhood = df_hospital_appts.groupby(['Gender', 'Neighbourhood'])
````

````python
group_by_gender_hipertension = df_hospital_appts.groupby(['Gender','Hipertension'])
````

We can find things like, 
* the average age of males vs females
* the number of males/females in each neighborhood
* the number of hypertensive or diabetic males vs females
* rank neigborhoods by highest numbers according to disease

#### Plotting Grouped data

````python
count_genders = group_by_gender.size()
````

````python
count_genders.plot(kind='barh',figsize=(16,6));
````

````python
group_by_gender_hipertension.size().unstack()
````

````python
count_gender_hipertension = group_by_gender_hipertension.size().unstack()
````

````python
count_gender_hipertension.plot(kind='bar', stacked=True, title='Hipertension - Males vs Females', ylabel='Count');
````
